#### **YELLOW TAXI**

The yellow taxi, often referred to as the "yellow cab," is an iconic symbol of New York City and is recognized worldwide as a symbol of the city's vibrant culture, energy, and urban lifestyle. The yellow taxi represents much more than just a mode of transportation, it embodies the spirit of NYC's fast-paced lifestyle, diversity, and the unique energy that makes the city so special. 

New York City is known for being a 24/7 city "City That Never Sleeps", and yellow taxis contribute to that image by being available around the clock to transport people to their destinations, day and night. **Yellow taxis are the only vehicles that are allowed to pick passengers up in response to a street hail across NYC**. But to legally operate a yellow cab in NYC, a driver must own a **medallion** issued by the TLC. 

![medallion](./medallion.png)

Medallion refers to a metal plate or tag that is affixed to the hood of a taxi cab, signifying that the vehicle has been licensed to operate as a taxi. The medallion system was established in 1937 by the NYC government to regulate the taxi industry and control the number of taxis on the road. For decades medallion price icreasing slowly over time. It peaked around 2013 at over 1 million USD. But when ride-sharing services like Uber and Lyft launched and flooded the supply of rides in NYC, the price had fallen for less than 138K USD. Since many cab drivers took out loans in order to afford medallions when values were high and then Uber and Lyft entered the industry and captured the market, they were drowning in debt and even take their own lives.

Nowadays, due to changes in the industry and rideshare services, many **drivers lease medallions** from medallion owners or from companies that manage fleets of taxis. This leasing system allows drivers to operate taxis without owning the medallion outright. So becoming a taxi driver in NYC someone must have a TLC Driver's License and Medallion and then drivers can legally pick up passengers by driving through the city, responding to street hails, and accepting pre-arranged trips.

With around 13K vehicles operating in NYC, the taxi industry is starting to recover since the COVID pandemic. Drivers are working and striving to meet their needs. Nowadays, there are more Uber and Lyft vehicles compared to yellow taxis. With regulations constantly changing, numerous additional fees, and especially with the presence of ridesharing like Uber and Lyft, taxi drivers as workers and community are actually facing a challenging situation. Taxi drivers are fighting for their place within the industry. **The latest news is that the Metropolitan Tax Authority (MTA) will raise the congestion surcharges from the previous 2.5 to 9-23 USD per trips starting in the spring of 2024.** "While passengers may be charged the additional surcharges, it is the drivers who will ultimately pay in significantly fewer trips and lower compensation.”

In this situation, drivers only can rely on TLC as a regulator, overseer and goverment agency to maintain competitiveness of the iconic symbol of NYC and ensure the sustanaibility both of the industry and the drivers. As TLC strategic plan in 2022 (https://www.nyc.gov/assets/tlc/downloads/pdf/taxi_strategic_plan_2022.pdf) Section 8 : Improving the Driver Experience, noted that **"Drivers are the core of the Taxi industry but they have reported difficult in attracting and maintaining the competitiveness of labor market. ### Missing Value
There are 6 columns having missing value (NaN) which are store_and_fwd_flag, RatecodeID, passenger_count, payment_type, trip_type, congestion_surcharge. We will handle this one by one.

1.  `congestion_surcharge`

Congestion surcharge is a surcharge for all trips that start, end, or pass through Manhattan south of 96th Street or Manhattan Central Business District. Typically implemented in designated zones to manage traffic congestion and generate revenue for transportation improvements. Total amount is the sum of fare_amount + extra + mta_tax + tip_amount + tolls_amount + improvement_surcharge + congestion_surcharge. Therefore null value in congestion_surcharge will be filled with the subraction of total_amount.

```python
# Replacing NaN in 'congestion_surcharge' with the the subraction of total_amount

df['congestion_surcharge'].fillna(df['total_amount'] - (df['fare_amount'] + df['extra'] + df['mta_tax'] + df['tip_amount'] + df['tolls_amount'] + df['improvement_surcharge']), inplace=True)
df['congestion_surcharge'] = df['congestion_surcharge'].clip(lower=0.0)
```

2.     `store_and_fwd_flag`

This column contains information about the trip record whether was held in the vehicle memory before sending to the vendor, aka “store and forward” or not. A "store and forward" trip in this context typically refers to a situation where the vehicle's data is stored on the vehicle's equipment due to a lack of real-time communication with the central server or dispatch system. This can happen when the vehicle is in an area with poor or no network connectivity, preventing immediate transmission of trip information to the central server.

When the vehicle eventually regains a stable network connection, the stored trip data is "forwarded" to the central server.

Y = store and forward trip,     N = not a store and forward trip

```python
# Checking the distribution of trip record type
trip_record_type = df['store_and_fwd_flag'].value_counts().reset_index().rename(columns={'index':'store_and_fwd_flag','store_and_fwd_flag':'count'})
trip_record_type['%'] = trip_record_type['count'] / len(df) * 100
trip_record_type
```

*   Nearly all of the trips are not a store and forward trips. Only 308 trips or equivalent to 0.45% of the data that store and forward. 
*   Taking this into consideration, we will fill the missing values in this column with 'N', given the observation that the vast majority of the data does not a store and forward trips.

```python
# Replacing NaN in store_and_fwd_flag
df.fillna({'store_and_fwd_flag':'N'}, inplace=True)
```

3.  `trip_type`

The "trip type" field with values "street hail" and "dispatch" refers to two different ways that passengers can request and initiate taxi trips. Here's the difference between the two:

*   Street hail
    Street Hail trips occur when passengers directly hail a taxi from the street. This is the traditional way of obtaining a taxi, where passengers stand on the side of the road and signal to an available taxi to pick them up.

*   Dispatch
    Dispatch trips involve the use of a centralized system to request a taxi. Passengers can call a taxi company or use a taxi-hailing app to request a taxi to a specific location. The request is then sent to a dispatcher, who assigns an available taxi to pick up the passenger. Dispatch trips can be pre-arranged or scheduled in advance.

```python
# Checking the distribution of trip type
# 1 = Street-hail 2 = Dispatch
taxi_trip_type = df['trip_type'].value_counts().reset_index().rename(columns={'index':'trip_type','trip_type':'count'})
taxi_trip_type['%'] = taxi_trip_type['count'] / len(df) * 100
taxi_trip_type
```

*   The dominant trip type is street hail while only 2% of the data in dispatch trip type.
*   Given this fact, the missing value in this column in this will be filled with '1' representing Street Hail type.

```python
# Replacing NaN in trip_type
df.fillna({'trip_type':1}, inplace=True)
```

4.  `RatecodeID`

This column contains information about the final rate code at the end of the trip. Rate code should be in range 1 up to 6.
1.  Standard rate : This trip type indicates a regular taxi trip within New York City that is subject to the standard fare structure.

2.  JFK : This trip type indicates that the trip originated from or is going to John F. Kennedy International Airport (JFK). 

3.  Newark : The trip originated from or is going to Newark Liberty International Airport. Similar to JFK trips.

4.  Nassau or Westchester : This trip type suggests that the trip was to or from Nassau County or Westchester County, which are suburban areas neighboring New York City.

5.  Negotiated fare :  A trip with a negotiated fare indicates that the fare for the trip was pre-negotiated or agreed upon between the driver and the passenger.

6.  Group ride : A trip classified as a group ride indicates that multiple passengers who are not together as a group are sharing the same ride.

```python
# Checking the distribution of rate code type.
rate_code_perc = df['RatecodeID'].value_counts().reset_index().rename(columns={'index':'RatecodeID','RatecodeID':'count'})
rate_code_perc['%'] = rate_code_perc['count'] / len(df) * 100
rate_code_perc
```

*   The dominant rate code is Standard Rate which indicates a regular taxi trip wihtin NYC, followed by Negotiated rate or RatecodeID 5.
*   Drivers must enter the total fare into the Taximeter as a flat rate trip charge using Rate Code ID 5 at the start of the Trip.
*   Given the missing values in this columns only 6% of the entire dataset, we choose to fill the missing value rather than delete them. Considering the nature of the dataset being about trips across NYC, we will fill the missing value with standard rate, indicating a regular taxi trip wihtin NYC.

```python
#Replacing NaN in RatecodeID
df.fillna({'RatecodeID':1}, inplace=True)
```

```python
# We will convert distance into four level

def distance_level(row):
    if (row > 0) & (row <= 2):
        val = "short_trips"
    elif (row > 2) & (row <= 5):
        val = "medium_trips"
    elif (row > 5) & (row <= 46):
        val = "long_trips"
    else:
        val = "too_long_trips"
    return val

df['trip_level'] = df['trip_distance'].apply(distance_level)
```

```python
def clock(ax, radii, title, color):
    N = 24
    bottom = 2

    # create theta for 24 hours
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)

    # width of each bin on the plot
    width = (2*np.pi) / N
    
    bars = ax.bar(theta, radii, width=width, bottom=bottom, color=color, edgecolor="#999999")

    # set the lable go clockwise and start from the top
    ax.set_theta_zero_location("N")
    # clockwise
    ax.set_theta_direction(-1)

    # set the label
    ax.set_xticks(theta)
    ticks = ["{}:00".format(x) for x in range(24)]
    ax.set_xticklabels(ticks)
    ax.set_title(title)

plt.figure(figsize = (12,12))
ax = plt.subplot(2,2,1, polar=True)
    # make the histogram that bined on 24 hour
radii = np.array(df['pu_hour'].value_counts(sort = False).tolist(), dtype="int64")
title = 'Yellow Taxi Trips per Hour'
clock(ax, radii, title, "#F8DE22")
```

### Adjusting Surcharges

These days the list of fees, taxes, and surcharges for riding in a taxi  has gotten long and complicated. Yellow Taxi includes 7 charge components : fare amount, extra, mta tax, tip, tolls, improvement surcharge, congestion surcharge. Fare amount, extra, tip and tolls go to the drivers while improvement surcharge, congestion surcharge and mta tax go to goverment. To be spesific about improving drivers income, let's examine the surcharges allocated to the government, as they **should remain fixed fares**.

*   $0.3 improvement surcharge refers to an additional charge that is added to the fare for taxi trips in New York City. This surcharge was introduced to support certain improvements and initiatives related to the taxi and for-hire vehicle industry and transportation infrastructure.

*   $2.5 congestion surcharge refers to a fee that is implemented to address traffic congestion in Manhattan areas.

*   $0.5 MTA Tax must be charged on any trip.

```python
# Checking mta tax surcharge
mta_tax = merge_df['mta_tax'].value_counts().reset_index().rename(columns= {'index':'mta_tax_surcharge','mta_tax':'number of trips'})
mta_tax
```

```python
# Checking congestion surcharge
merge_df['congestion_surcharge'] = merge_df['congestion_surcharge'].astype(float)
def format_value(x):
    return "{:.2f}".format(x)
merge_df['congestion_surcharge'] = merge_df['congestion_surcharge'].apply(format_value)
congestion = merge_df['congestion_surcharge'].value_counts().reset_index().rename(columns= {'index':'congestion surcharge','congestion_surcharge':'number of trips'})
congestion
```

```python
# Checking improvement surcharge
improvement = merge_df['improvement_surcharge'].value_counts().reset_index().rename(columns= {'index':'improvement surcharge','improvement_surcharge':'number of trips'})
improvement
```

```python
figure, axes = plt.subplots(1,3, figsize=(8,5))
colors = ['#F8DE22']
sns.barplot(data=mta_tax, x ='mta_tax_surcharge', y='number of trips', palette = colors, ax= axes[0])
axes[0].set_title('MTA TAX')
sns.barplot(data=congestion, x ='congestion surcharge', y='number of trips', palette = colors, ax= axes[1])
axes[1].set_title('Congestion')
sns.barplot(data=improvement, x ='improvement surcharge', y='number of trips', palette = colors, ax= axes[2])
axes[2].set_title('Improvement')
plt.tight_layout()
plt.show()
```

These charts clarify that all surcharges that are designated for the government are not fixed fares. If the surcharges were set to zero, it would imply that taxi trips would not include surcharges as part of the fare calculation. This could have several potential implications, such as promotional offers, impacts on MTA funding, or for zero congestion charge indicating that the trip doesn't start, end, or pass through the Manhattan central business district. The MTA tax should be 0.5 USD per trip, the congestion surcharge should be 2.5 USD, and the improvement surcharge should be 0.3 USD. However, in the dataset dictionary, the improvement surcharge is listed as 0.3 USD, while on the TLC website, it's listed as 1 USD. TLC should maintain consistency in all these prices. Imagine if the fare went to drivers for tips or extra charges instead.
